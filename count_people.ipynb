```
import cv2 
import imutils
import numpy as np
from google.colab.patches import cv2_imshow
```

```
class People(object):
    def __init__(self, img_path):
        self.img_path = img_path
        self.counter = 0 
        self.ppl_img = []
        self.image = None
        
        
    def load_files(self): 
      '''
      '''
      im = cv2.imread(self.img_path, 0) #load as grey
      face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
      self.image = im
     # cv2_imshow(im)
      #cv2.waitKey(0)
      #cv2.destroyAllWindows()

        
      return face_cascade
    
    def detect_people(self):
      '''
      1- Using a classifier ( Haarcascade ) to detect people
      A) Convert the image into gray scale
      B) Load the classifier
      C) Detect Faces with the gray image, scale factor, and minNeighbours 
      '''
      face_detected = self.load_files().detectMultiScale(image = self.image, scaleFactor = 1.1, minNeighbors = 3 )
      
      return face_detected 
      

    def count_people(self):
      '''
      '''

      new = self.detect_people() #new result
      #imm = np.array()
      for (x, y,w,h) in new:
        imm = cv2.rectangle(self.image, (x,y), (x+w,y+h), (0,0,255), 3)
      cv2_imshow(imm)
      return len(new) 
      

      
        
    
```
